```python
from IPython.core.interactiveshell import InteractiveShell
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
import sympy as sp
import sympy.physics.mechanics as me

np.set_printoptions(precision=2, suppress=True)
%precision 2

InteractiveShell.ast_node_interactivity = "all"

me.init_vprinting()
```

```python
t = sp.symbols('theta')
h, b1, b2, b3, b4, n = 5*sp.cos(6*sp.pi/180), 55, 5, 55, 245, 1000
s1 = h * (t/b1 - 1/sp.pi/2*sp.sin(2*sp.pi*t/b1))
s2 = s2n = np.ones(int(n*b2/360))*h
s3 = -s1 + h
s4 = s4n = np.zeros(int(n*b4/360))
s1n = sp.lambdify(t,s1)
s3n = sp.lambdify(t,s3)
t1 = np.linspace(0,b1,int(n*b1/360),endpoint=0)
t2 = np.linspace(0,b2,int(n*b2/360),endpoint=0)
t3 = np.linspace(0,b3,int(n*b3/360),endpoint=0)
t4 = np.linspace(0,b4,int(n*b4/360))
# t1+t2+t3+t4
xs = np.hstack([t1,b1+t2,b1+b2+t3,b1+b2+b3+t4])
ys = np.hstack([s1n(t1),s2n,s3n(t3),s4n])
fig = go.Figure()
fig.add_trace(go.Scatter(x=xs,y=ys,name='S'))
fig.show()
```

```python
ds1 = s1.diff()
ds2 = ds2n = np.zeros(int(n*b2/360))
ds3 = s3.diff()
ds1n = sp.lambdify(t, ds1)
ds3n = sp.lambdify(t, ds3)
plt.grid()
plt.plot(t1,ds1n(t1),b1+t2,ds2n,b1+b2+t3,ds3n(t3),b1+b2+b3+t4,s4n)
```

```python
dds1 = ds1.diff()
dds3 = ds3.diff()
dds1n = sp.lambdify(t, dds1)
dds3n = sp.lambdify(t, dds3)
plt.grid()
plt.plot(t1,dds1n(t1),b1+t2,ds2n,b1+b2+t3,dds3n(t3),b1+b2+b3+t4,s4n)
```

```python
ddds1 = dds1.diff()
ddds3 = dds3.diff()
ddds1n = sp.lambdify(t, ddds1)
ddds3n = sp.lambdify(t, ddds3)
plt.grid()
plt.plot(t1,ddds1n(t1),b1+t2,ds2n,b1+b2+t3,ddds3n(t3),b1+b2+b3+t4,s4n)
```
