```python
# %load preload.py

from matplotlib import rc

import matplotlib.pyplot as plt
from numpy import *
from control import *
from scipy.linalg import *
from sympy.solvers.solveset import linsolve
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
# from sympy import Matrix, symbols, sin, cos, tan, asin, acos, atan, simplify
from scipy.integrate import solve_ivp

plt.rc('axes',axisbelow=True)
rc('text', usetex=True)
init_vprinting()
```

```python
F = tf([1,1.5],poly([0,-1,-10]))
x = linspace(-10,10,10000)
r, k = root_locus(F)
damping_ratio = tan(pi-arccos(.7998396910782286)) * x
plt.plot(x, damping_ratio)
plt.savefig('root.png',dpi=600)
```

```python
drline = asarray(x,dtype=complex)
drline.imag = damping_ratio
```

```python
interval = (-6,-4)
x1 = where((interval[0]<real(drline)) & (real(drline)<interval[1]),drline,False)
x2 = where((interval[0]<real(r)) & (real(r)<interval[1]) & (imag(r)!=0), r, False)[:,1]
```

```python
xs = asarray([i for i in x1 if i!=0])
ys = asarray([i for i in x2 if i!=0])
```

```python
for y in ys:
    if min(abs(xs-y)) > 0.2: continue
    print(k[list(x2).index(y)])
    print(min(abs(xs-y)), y)
```
